```python
# import numpy as np 
import pandas as pd
```

```python
litreracy = pd.read_csv('2015_16_Statewise_Secondary.csv')
```

```python
litreracy.head()
```

```python
column = litreracy.columns[0:18]
rename_col = column.str.strip().str.lower()
dict = {}
j=0
for i in column:
    dict[i] = rename_col[j]
    j+=1
```

```python
litreracy.rename(dict, inplace = True)
lit_new = litreracy.iloc[:,0:18]
lit_new.head()
```

```python
lit_new = lit_new[lit_new['literacy_rate'].notnull()]
lit_new.isnull().sum()
```

```python
lit_new['statname'].value_counts().sum()
```

```python
lit_new.duplicated().sum()
```

No duplicate rows are present in the dataset

```python
%matplotlib inline
lit_new_sort = lit_new.sort_values('literacy_rate')
lit_new_sort.plot(x = 'statname', y = 'literacy_rate', kind = 'bar',figsize = (18,10),rot = 35b)
```

Clealry, we can see that kerela has highest while bihar has lowest literacy rate
